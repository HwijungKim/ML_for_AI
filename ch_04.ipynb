```python
weather = pd.read_csv('data/weather_by_station.csv', index_col='date', parse_dates=True)
weather.head()
```

```python
dp = weather.query('datatype == "PRCP"').groupby('date')['value'].sum()
dp = pd.DataFrame(dp)
dp 
```

```python
w1 = dp.groupby(pd.Grouper(freq='Q')).mean()
w1
```

```python
dp = dp.reset_index()
dp
```

```python
dp['month'] = dp['date'].dt.month
dp
```

```python
days_in_month = dp.groupby('month')['date'].nunique().reset_index()
days_in_month.columns = ['month', 'days_in_month']
dp = pd.merge(dp, days_in_month, on='month', how='left')
dp
```

```python
dp['daily_prcp_normalized'] = dp['value'] / dp['days_in_month']
dp
```

```python
dp.set_index('date', inplace=True)
dp['value'].rolling(window='3D').mean()
```

```python
fb = fb = pd.read_csv('data/fb_2018.csv', parse_dates=True).assign(
    trading_volume=lambda x: pd.cut(x.volume, bins=3, labels=['low', 'med', 'high'])
)
fb
```

```python
fb['date'] = pd.to_datetime(fb['date'])
```

```python
fb['month'] = fb['date'].dt.month
```

```python
fb.set_index('date', inplace=True)
```

```python
result = fb.pivot_table(
    index='trading_volume',
    columns=fb.index.month,
    aggfunc='size',
    fill_value=0
)

result.columns.name = 'month'
```
